```python
days=["mon", "tue","wed","thu","fri","sat","sun"]
temps=[22,24,21,23,25,27,26]

plt.figure(figsize=(10,5))
plt.plot(days,temps,marker='o', linestyle='-',color='red')
plt.title('Weekly Temperature trends')
plt.xlabel('Day of the week')
plt.ylabel('Temperature (C)')
plt.grid(True)
plt.show()
```

```python
hours_studied=[1,2,3,4,5,6,7,8,9]
test_scores=[55,60,65,65,70,70,75,76,80]

plt.scatter(hours_studied,test_scores)
plt.title('Scatter plot of hours studied vs. Test scores')
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')
plt.show()
```
